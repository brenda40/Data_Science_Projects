```python
# Brenda Woodard
# Data Visualization 
# Assignment 2 

import wget
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, date, time, timedelta
```

```python
# Assignment 2 Part 1

A_W = [1, 4, 7, 9, 11, 12]
labels=['Hydrogen','Helium','Lithium','Beryllium', 'Boron', 'Carbon']
sizes= [1, 4, 7, 9, 11, 12] 
explode = (0,0,0,0,0,0.1)
explode_2 = (0.4,0,0,0,0,0)

fig,ax = plt.subplots(1, 2, figsize =(14,6), num=59)

ax[0].pie(sizes, labels =labels, explode= explode,
          autopct='%1.1f%%', shadow = True, colors = ['blue', 'lightblue', 'darkgreen', 'lightgreen', 'purple', 'violet'])
ax[0].set_title('Atomic Weights')
ax[0].axis('equal')

ax[1].pie(sizes, labels =labels, explode = explode_2,
          autopct =lambda p: '{:.0f}'.format(p * sum(sizes) / 100.0), shadow = True, colors = ['blue', 'lightblue', 'darkgreen', 'lightgreen', 'purple', 'violet'])
ax[1].set_title('Atomic Weights')
ax[1].axis('equal')
plt.show()
```

```python
# Assignment 2 Part 2

IDE_data = pd.read_csv("py_ide2.csv")
IDE_df = pd.DataFrame(IDE_data)

fig, ax = plt.subplots(1, 2, figsize =(14,8), num =91)
ax[0].barh(IDE_df['IDE'], IDE_df['Adoption'], color='lightblue',  edgecolor='blue')
ax[0].set_title('IDE 2')
ax[0].set_xlabel('Adoption')
ax[0].set_ylabel('IDE')
plt.xticks(range(len(IDE_df)), IDE_df['IDE'],
           rotation=35)

ax[1].bar( range(len(IDE_df)), IDE_df['Adoption'],  color='lightblue', edgecolor='blue')
ax[1].set_title('IDE 2')
ax[1].set_ylabel('Adoption')
ax[1].set_xlabel('IDE')
plt.tight_layout()
plt.show()
```

```python
# Assignment 2 Part 3

dates = ['2012-12-16', '2012-10-16','2012-08-16','2012-06-16', '2012-04-16','2012-02-16', '2011-12-16', '2011-10-16']
Array = np.random.uniform(100,200,8)

D_df = {'date':dates, 'array':Array}
D2_df = pd.DataFrame(D_df)

D2_df['date'] = pd.to_datetime(D2_df['date'])
index_D2_df = D2_df.set_index('date')
values = index_D2_df['array']


plt.figure(num= 100, figsize= (16,8))

plt.subplot(211)
plt.plot(dates, values, color = 'g')
plt.title('Array Values')
plt.ylabel('Values')
plt.xticks(rotation = -20)

plt.subplot(212)
plt.bar(dates, values,  color= 'g')
plt.ylabel('Values')
plt.xlabel('Date')
plt.xticks(rotation = -20)

plt.tight_layout()
plt.show()
```

```python
# Assignment 2 Part 4

stock = pd.read_csv('BCN-USD.csv', delimiter = ',')
s_df = pd.DataFrame(stock)
s_df_trim = s_df[330:]

stock['Date'] = pd.to_datetime(stock['Date'])
index_stock = stock.set_index('Date')
close_price = index_stock['Close']
volume = index_stock['Volume']

plt.figure(num= 100, figsize= (16,8))

plt.subplot(211)
plt.plot(close_price, color = 'r')
plt.title('Bitcoin')
plt.ylabel('Closing Price')
plt.xticks(rotation = 20)

plt.subplot(212)
plt.plot(volume, color = 'r')
plt.ylabel('Volume')
plt.xlabel('Date')
plt.xticks(rotation = 20)
plt.tight_layout()
plt.show()
```
